```python
import base64
import os

def make_base64_file_payload(file_path: str):
    """
    Read a file and return (base64_string, file_name)
    to simulate what Streamlit sends to the API.
    """
    with open(file_path, "rb") as f:
        file_bytes = f.read()
    encoded = base64.b64encode(file_bytes).decode("utf-8")
    file_name = os.path.basename(file_path)
    return encoded, file_name
```

```python
from wulfs_routing_api.utils.data_io_utils import base64_data_to_temp_file
daily_orders_path = "../../data/daily_orders.xlsx" 
encoded_data, file_name = make_base64_file_payload(daily_orders_path)
temp_file_path = base64_data_to_temp_file(file_name,encoded_data)
temp_file_path
```
