```python
import math
def compute_haversine_distance(coord1, coord2):
    """Compute great-circle distance (km) between two lat/lon coordinates."""
    R = 6371.0  # Earth radius in km
    lat1, lon1 = map(math.radians, coord1)
    lat2, lon2 = map(math.radians, coord2)
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    return 2 * R * math.asin(math.sqrt(a))
```

```python
locations = [(row['Latitude'], row['Longitude']) for _, row in wulfs_cust_df.head(5).iterrows()]
depot = (42.3490175,-71.0311327)
locations.insert(0, depot)

for location in locations:
    distance = compute_haversine_distance(depot,location)
    print(f"{location} {distance}")
```
